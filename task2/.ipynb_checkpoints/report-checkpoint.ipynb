## Эксперимент 5. 3-fold кросс-валидация 

Ниже переписано содержимое файла cross_validation.py ради использования анимированных прогресс-баров, взятых с гитхаба alexanderkuk

```python
def log_progress(sequence, every=None, title='AZAZA', size=None):
    from ipywidgets import IntProgress, HTML, VBox
    from IPython.display import display

    is_iterator = False
    if size is None:
        try:
            size = len(sequence)
        except TypeError:
            is_iterator = True
    if size is not None:
        if every is None:
            if size <= 200:
                every = 1
            else:
                every = size / 200     # every 0.5%
    else:
        assert every is not None, 'sequence is iterator, set every'

    if is_iterator:
        progress = IntProgress(min=0, max=1, value=1)
        progress.bar_style = 'info'
    else:
        progress = IntProgress(min=0, max=size, value=0)
    label = HTML()
    box = VBox(children=[label, progress])
    display(box)

    index = 0
    try:
        for index, record in enumerate(sequence, 1):
            if index == 1 or index % every == 0:
                if is_iterator:
                    label.value = title+':{index} / ?'.format(index=index)
                else:
                    progress.value = index
                    label.value = title+u':{index} / {size}'.format(
                        index=index,
                        size=size
                    )
            yield record
    except:
        progress.bar_style = 'danger'
        raise
    else:
        progress.bar_style = 'success'
        progress.value = index
        label.value = str(index or '?')
        progress.close()
        label.close()
        
    box.close()
```

```python

def kfold(n, n_folds=3):
    idx = np.arange(n)
    fold = np.array([])
    fold_sz = int(n/n_folds)

    test_indices = []
    train_indices = []

    for fold_num in range(n_folds):
        fold = idx[fold_num*fold_sz:(fold_num+1)*fold_sz]
        test_indices.append(fold)
        train_indices.append(idx[~np.in1d(idx, fold)])

    return list(zip(train_indices, test_indices))


def get_accuracy(pred, real):
    real = list(real)
    pred = list(pred)
    my_sum = 0.0
    for idx in range(len(pred)):
        my_sum += float(pred[idx] == real[idx])
    return my_sum/len(real)


def get_fscore(pred, real):
    classes = np.unique(real)
    f_score = 0.0

    for class_tag in classes:
        p_tag = (pred == class_tag)
        r_tag = (real == class_tag)

        TP = np.logical_and(p_tag, r_tag).sum()
        if TP < 0.0001:  # it means that precision and recall equal to 0
            continue

        FP = np.logical_and(p_tag, np.logical_xor(p_tag, r_tag)).sum()
        FN = np.logical_and(r_tag, np.logical_xor(p_tag, r_tag)).sum()
        precision = TP / (TP+FP)
        recall = TP / (TP+FN)
        f_score += 2*precision*recall/(precision+recall)

    return f_score/len(classes)


def knn_cross_val_score(X, y, k_list, score, cv=None, weights=False, 
                        metric="euclidean", strategy="brute", 
                        test_block_size=1000):
    if cv is None:
        cv = kfold(len(X))
    k_list = sorted(k_list)
    folds_num = len(cv)

    my_knn = KNN_classifier(k_list[-1], strategy, metric, weights,
                            test_block_size, True)

    k_scores = {}
    for k in k_list:
        k_scores[k] = np.zeros(folds_num)

    if score == "accuracy":
        for fold in log_progress(range(folds_num), every=1, title="folds"):
            my_knn.fit(X[cv[fold][0]], y[cv[fold][0]])
            pred = my_knn.predict(X[cv[fold][1]])
            k_scores[k_list[-1]][fold] = get_accuracy(pred, y[cv[fold][1]])

            for k in k_list[:-1]:
                pred = my_knn.predict(X[cv[fold][1]], k)
                k_scores[k][fold] = get_accuracy(pred, y[cv[fold][1]])
        return k_scores
    elif score == "f_score":
        for fold in log_progress(range(folds_num), every=1, title="folds"):
            my_knn.fit(X[cv[fold][0]], y[cv[fold][0]])
            pred = my_knn.predict(X[cv[fold][1]])
            k_scores[k_list[-1]][fold] = get_fscore(pred, y[cv[fold][1]])

            for k in k_list[:-1]:
                pred = my_knn.predict(X[cv[fold][1]], k)
                k_scores[k][fold] = get_fscore(pred, y[cv[fold][1]])
        return k_scores
```

## Эксперимент 8. Построение confusion matrix

```python
predictions = np.array(predictions)

wrong_pics = test_data[np.where(test_labels != predictions)]
wrong_real = test_labels[np.where(test_labels != predictions)]
wrong_pred = predictions[np.where(test_labels != predictions)]
wrong_len = len(wrong_pics)
plt.figure(figsize=(12, 12))

for i in range(10):
        rand_ind = random.randint(0, wrong_len-1)
        while wrong_real[rand_ind] != i:
            rand_ind = random.randint(0, wrong_len-1)
        plt.subplot(10, 5, i+1)
        plt.axis("off")
        plt.title("Real:"+str(wrong_real[rand_ind]) + "  Pred:" + str(wrong_pred[rand_ind]))
        plt.imshow(wrong_pics[rand_ind].reshape((28, 28)), cmap="Greys");
```

```python
from sklearn.metrics import confusion_matrix
import itertools
# code was taken from sklearn examples of plotting confusion matrices
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.cool):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
#         print("Normalized confusion matrix")
#     else:
#         print('Confusion matrix, without normalization')

#     print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, np.around(cm[i, j], decimals=5),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
my_confusion = confusion_matrix(test_labels, predictions)

# Plot non-normalized confusion matrix
plt.figure(figsize=(12, 12))
plot_confusion_matrix(my_confusion, classes=np.arange(10),
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure(figsize=(12, 12))
plot_confusion_matrix(my_confusion, classes=np.arange(10).astype(np.int64),
                      normalize=True, title='Normalized confusion matrix')

plt.show()
```

```python
for i in range(len(news20_dataset.data)):
    news20_dataset.data[i] = list(filter(None, pattern.split(news20_dataset.data[i])))
```

```python
print(news20_dataset.data[0])
```

## Эксперимент 12. Разреженная матрица

```python
from scipy.sparse import csr_matrix
texts_num = len(news20_dataset.data)
csr_indices = []
csr_data = []
csr_indptr = [0]
total_words_list = []

for text in news20_dataset.data:
    total_words_list.extend(text)
    csr_indptr.append(csr_indptr[-1]+len(text))

words_list = list(set(total_words_list))
words_num = len(words_list)
words_dict = {}

for idx, word in enumerate(words_list):
    words_dict[word] = idx
    
csr_data = [1 for i in range(len(total_words_list))]
for text in news20_dataset.data:
    for word in text:
        csr_indices.append(words_dict[word])
news20_csr = csr_matrix((csr_data, csr_indices, csr_indptr), 
                        shape=(texts_num, words_num))
```

## Эксперимент 13. Преобразование tf-idf

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_class = TfidfTransformer()

news20tfidf_csr = tfidf_class.fit_transform(news20_csr)
print(news20tfidf_csr.shape)
```
