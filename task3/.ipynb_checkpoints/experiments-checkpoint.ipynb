```python
import numpy as np
import scipy as sc
import sklearn.utils as skutils
import svm
import matplotlib.pyplot as plt
%matplotlib inline
```

# Генерация выборки

```python
X_fst = np.random.randn(100, 2)
X_sec = 5*np.random.randn(100, 2)+20
X_train = np.vstack((X_fst, X_sec))
y_train = np.hstack((np.ones(100), np.zeros(100)-1))
X_train, y_train = skutils.shuffle(X_train, y_train)
```

```python
print(y_train)
```

```python
X_fst = np.random.randn(20, 2)
X_sec = 5*np.random.randn(20, 2)+20
X_test = np.vstack((X_fst, X_sec))
y_test = np.hstack((np.ones(20), np.zeros(20)-1))
X_test, y_test = skutils.shuffle(X_test, y_test)
```

# Тестирование 'liblinear'

```python
import imp

imp.reload(svm)
my_svm = svm.SVM(C=0.2, method='liblinear')
my_svm.fit(X_train, y_train)
y_pred = my_svm.predict(X_test)
```

```python
print("Primal objective:", my_svm.compute_primal_objective(X_test, y_test))
```

```python
def get_accuracy(pred, real):
    real = list(real)
    pred = list(pred)
    my_sum = 0.0
    for idx in range(len(pred)):
        my_sum += float(pred[idx] == real[idx])
    return my_sum/len(real)

print("Accuracy:", get_accuracy(y_pred, y_test))
```

```python
svm.visualize(X_train, y_train, my_svm)
```

# Testing 'subgradient'

```python
import random
X_gleb = np.zeros((40, 2))
for i in range(20):
    X_gleb[i, 0] = random.random()*5
    X_gleb[i, 1] = random.random()*3 + 1
    X_gleb[i+20, 0] = (random.random()+1)*5
    X_gleb[i+20, 1] = (random.random()+0.5)*10

y_gleb = np.zeros(40)
for i in range(20):
    y_gleb[i] = 1
    y_gleb[i+20] = -1
```

```python
my_svm = svm.SVM(C=1.0, method='subgradient')
my_svm.fit(X_train, y_train, tol=1e-4, max_iter = 10000, verbose=False,
           stop_criterion="argument", alpha=0.5, beta=1.0)
```

```python
plt.figure(figsize=(6, 6))
svm.visualize(X_train, y_train, my_svm)
```

```python
my_svm = svm.SVM(C=1.0, method='subgradient')
my_svm.fit(X_train, y_train, tol=1e-4, max_iter = 10000, verbose=False,
           stop_criterion="objective", alpha=0.5, beta=1.0)
```

```python
plt.figure(figsize=(6, 6))
svm.visualize(X_train, y_train, my_svm)
```

# Testing stochastic gradient

```python
my_svm = svm.SVM(C=0.5, method='stoch_subgradient')
my_svm.fit(X_train, y_train, tol=1e-5, max_iter = 10000, verbose=False,
           stop_criterion="objective", alpha=0.5, beta=1.0)
```
